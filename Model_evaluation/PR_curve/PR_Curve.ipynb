```python
import joblib
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
```

```python
XGB = joblib.load('XGB_Final.pkl')
SVM = joblib.load('SVM_Final.pkl')
LR = joblib.load('LR_Final.pkl')
NB = joblib.load('NB_Final.pkl')
DT = joblib.load('DT_Final.pkl')
KNN = joblib.load('KNN_Final.pkl')
```

```python
test_data = pd.read_csv('SHP2_test_descriptors_150.csv')
test_data
```

```python
final_features = ['MaxPartialCharge', 'FpDensityMorgan2', 'BCUT2D_CHGLO', 'BCUT2D_MRHI',
       'PEOE_VSA12', 'PEOE_VSA6', 'SMR_VSA3', 'SlogP_VSA3', 'SlogP_VSA8',
       'EState_VSA6', 'NumHAcceptors', 'NumSaturatedCarbocycles',
       'fr_bicyclic', 'TARGET']
```

```python
test_data = test_data[final_features]
test_data
```

```python
X_test = test_data.iloc[:,:-1].values
y_test = test_data.iloc[:,-1].values
```

```python
plt.figure(figsize=(10,10), dpi=1200)
# XGB
y_pred1 = XGB.predict_proba(X_test)[:, 1]
precision1, recall1, thresholds1 = metrics.precision_recall_curve(y_test, y_pred1)
# print(len(precision1))
plt.plot(precision1, recall1, label="XGBOOST")

# # # SVM
# y_pred2 = SVM.predict_proba(X_test)[:, 1]
# precision2, recall2, thresholds2 = metrics.precision_recall_curve(y_test, y_pred2)
# plt.plot(precision2, recall2, label="SVM")

# # # LR
# y_pred3 = LR.predict_proba(X_test)[:, 1]
# precision3, recall3, thresholds3 = metrics.precision_recall_curve(y_test, y_pred3)
# plt.plot(precision3, recall3,label="LR")

# # # NB
# y_pred4 = NB.predict_proba(X_test)[:, 1]
# precision4, recall4, thresholds4 = metrics.precision_recall_curve(y_test, y_pred4)
# plt.plot(precision4, recall4,label="NB")

# # DT
y_pred5 = DT.predict_proba(X_test)[:, 1]
precision5, recall5, thresholds5 = metrics.precision_recall_curve(y_test, y_pred5)
plt.plot(precision5, recall5,label="DT")

# # KNN
y_pred6 = KNN.predict_proba(X_test)[:, 1]
precision6, recall6, thresholds6 = metrics.precision_recall_curve(y_test, y_pred6)
plt.plot(precision6, recall6,label="KNN")

plt.rcParams.update({'font.size': 18})
plt.title('Precision-Recall Curve')
plt.xlabel("Recall")
plt.ylabel("Precision")

# #add legend
plt.legend(loc='lower left')
plt.savefig('..\images\PR_Curve.png')
```

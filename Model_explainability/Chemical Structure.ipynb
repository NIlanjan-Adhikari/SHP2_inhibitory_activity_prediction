```python
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
smiles = 'O=C(C(C=C1O)=CC2=CC=C(O)C(O)=C2C1=O)O'
smiles = 'O=C(C(C=C1O)=CC2=CC=C(O)C(O)=C2C1=O)O' # index 0
m = Chem.MolFromSmiles(smiles)
Draw.MolToImage(m)
```

```python
smiles = 'O=C(C(C=C1O)=CC2=CC=C(O)C(O)=C2C1=O)OC(C)C'
smiles = 'O=C(C(C=C1O)=CC2=CC=C(O)C(O)=C2C1=O)OC(C)C' # index 3
m = Chem.MolFromSmiles(smiles)
Draw.MolToImage(m)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Final_data/SHP2_train_descriptors_2741.csv')
data
```

```python
def activity(s):
    if s[0] == '<':
        if float(s[1:]) <= 1:
            return "Strong Inhibitor"
        else:
            return np.nan
    elif s[0] == '>':
        if float(s[1:]) >= 1:
            return "Weak Inhibitor"
        else:
            return np.nan
    else:
        if float(s) > 1:
            return "Weak Inhibitor"
        else:
            return "Strong Inhibitor"
```

```python
data['TARGET'] = data['IC50(microM)'].apply(activity)
data
```

```python
final_features = ['MaxPartialCharge', 'FpDensityMorgan2', 'BCUT2D_CHGLO', 'BCUT2D_MRHI',
       'PEOE_VSA12', 'PEOE_VSA6', 'SMR_VSA3', 'SlogP_VSA3', 'SlogP_VSA8',
       'EState_VSA6', 'NumHAcceptors', 'NumSaturatedCarbocycles',
       'fr_bicyclic', 'TARGET']
```

```python
data = data[final_features]
data
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data.iloc[:,:-1].corr(), annot=True)
plt.savefig(".\images\correlation_matrix.png")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

```python
data = pd.read_csv('..\Data_preprocessing\SHP2_train_descriptors_2741.csv')
data
```

```python
def activity(s):
    if s[0] == '<':
        if float(s[1:]) <= 1:
            return 1
        else:
            return np.nan
    elif s[0] == '>':
        if float(s[1:]) >= 1:
            return 0
        else:
            return np.nan
    else:
        if float(s) > 1:
            return 0
        else:
            return 1
```

```python
data['TARGET'] = data['IC50(microM)'].apply(activity)
data
```

# Creating 10 folds

```python
data["Kfold"] = -1
data = data.sample(frac=1).reset_index(drop=True)
y = data.TARGET.values
kf = StratifiedKFold(n_splits=10, shuffle=True)

for f, (t_, v_) in enumerate(kf.split(X=data, y=y)):
    data.loc[v_, 'Kfold'] = f

data
```
